```python
import pandas as pd
import matplotlib.pyplot as plt

mnist = pd.read_csv("mnist.csv")
label = mnist['label']
data = mnist.drop("label", axis=1)
print(label[10])
```

```python
#Seeing the image
plt.figure(figsize=(8,8))
index = 33
print(label[index])
grid_data = data.iloc[index].as_matrix().reshape(28,28)
plt.imshow(grid_data, interpolation=None, cmap="gray")
plt.show()
```

```python
#Doing Column Standardization
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
```

```python
#Finding Covariance matrix
import numpy as np
sample_data = standardized_data
covariance_matrix = np.matmul(sample_data.T, sample_data)
```

```python
from scipy.linalg import eigh
values, vectors = eigh(covariance_matrix, eigvals=(782,783))
#here we have two vectors,
#First one corresponds to First principle component analysis
#Second is 2nd prniciple component analysis
```

```python
new_coordinates = np.matmul(sample_data, vectors)
label = label.reshape(42000,1)
new_coordinates = np.append(new_coordinates, label, axis=1)
dataframe = pd.DataFrame(data=new_coordinates, columns=['1st_component', '2nd_component',' label'])
```

```python
import seaborn as sns
sns.FacetGrid(dataframe, hue=dataframe.columns[2], height=6).map(plt.scatter, "1st_component", "2nd_component").add_legend()
plt.show()
```

```python
#Using ScikitPCA directly
import seaborn as sns
from sklearn import decomposition
pca = decomposition.PCA(n_components=2)
pca_data = pca.fit_transform(sample_data)
pca_data = np.append(pca_data, label, axis=1)
dataframe1 = pd.DataFrame(data=pca_data, columns=['1st_component', '2nd_component', 'label'])

sns.FacetGrid(dataframe1, hue=dataframe1.columns[2], height=6).map(plt.scatter, "1st_component", "2nd_component").add_legend()
plt.show()
```

```python
iris = pd.read_csv("iris.csv")
label = iris['species']
data = iris.drop('species', axis=1)
sample_data = StandardScaler().fit_transform(data)
label = label.as_matrix().reshape(150,1)

pca = decomposition.PCA(n_components=2)
pca_data = pca.fit_transform(sample_data)

pca_data = np.append(pca_data, label, axis=1)
dataframe = pd.DataFrame(data=pca_data, columns=['1st', '2nd', 'species'])

sns.FacetGrid(dataframe, hue='species', height=6).map(plt.scatter, '1st', '2nd').add_legend()
plt.show()
```
