```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

mnist = pd.read_csv("mnist.csv")
sample_data = mnist[:1000]
label = sample_data['label']
data = sample_data.iloc[:,1:]

from sklearn.manifold import TSNE
model = TSNE(n_components=2, random_state=0 )
tsne_data = model.fit_transform(data)
```

```python
import seaborn as sns

label = label.as_matrix().reshape(1000,1)
tsne = np.append(tsne_data, label, axis=1)
df = pd.DataFrame(tsne, columns=['D1', 'D2', 'label'])
sns.FacetGrid(data=df, hue='label', height=6).map(plt.scatter, "D1", "D2").add_legend()
plt.show()
```
